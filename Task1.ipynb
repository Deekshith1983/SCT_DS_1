```python
countries = ["India","China","United States","Brazil","Indonesia"]
filtered_df = df[df["Country Name"].isin(countries)]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(filtered_df['Country Name'],filtered_df["2020"].astype(float))
plt.title("Population of selected countries in 2020")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Drop rows with no 2020 population data
df_cleaned = df[df["2020"].notna()]

# Convert 2020 population to numeric
populations = df_cleaned["2020"].astype(float)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(populations, bins=30, edgecolor='black')
plt.title("Distribution of Country Populations in 2020")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
# Get India's population over the years
india = df[df["Country Name"] == "India"]

# Extract years and population values
years = [str(year) for year in range(1960, 2024)]
population = india[years].values.flatten().astype(float)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(years, population, marker='o')
plt.title("India's Population Over Time (1960–2023)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Choose any 10 countries
top_10 = df[['Country Name', '2023']].dropna().head(10)
top_10['2023'] = top_10['2023'].astype(float)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10['Country Name'], top_10['2023'])
plt.title("Population of First 10 Countries in Dataset (2023)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Clean data
df_cleaned = df[df["2023"].notna()]
df_cleaned["2023"] = df_cleaned["2023"].astype(float)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned["2023"], bins=30, edgecolor='black')
plt.title("Distribution of Populations Across Countries (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset with age and gender
data = {
    'Age': [23, 25, 29, 30, 22, 28, 32, 40, 22, 23, 35, 38, 27, 28, 30],
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}
df = pd.DataFrame(data)
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
